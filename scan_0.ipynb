```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
x=pd.read_csv('data.csv')
```

```python
x=np.array((x))
```

```python
m=len(x[:,1])
m
```

```python
m=len(x[:,1])
n_batchs = m/761
n_batchs=int(n_batchs)
n_batchs
```

```python
def decision (xx):  # input argument is the mini batch of x
  c=[]     # c is a list that has the if condition results [block:0,pass:1,free:2]
  ind=[]     # ind is a list that has the indices of the center point that dron can access 55/2=27 
  for t in range(len(xx)-55):
    avr=np.mean(xx[t:t+55])
    if avr < 2.5:
      # print("Block")
      c.append(0)  
    elif avr <= 3:
      # print("Pass")
      c.append(1)
      ind.append(t+27)
    else:
      # print("free")
      c.append(2)
      ind.append(t+27)
    t=t+1
  return c,ind
```

```python
for k in range(n_batchs):
  x_batch = x[k*761:k*761+761,1]
  c,ind=decision (x_batch)
  print(len(ind))


  # setting the axes projection as polar 
  plt.axes(projection = 'polar') 
  
# setting the radius (original data)
  r = x_batch
  theta= np.arange(-(2 * np.pi/6), (2 * np.pi/6),2*np.pi/(3*761))


# plotting the original data  
  for i,j in zip(theta,r): 
    plt.polar(i, j, 'g.')
# ploting the angle that drone can move on
  r_pass=[]
  theta_pass=[]
  for o in range(len(ind)):         
    [ r_pass.append(5) for p in range (len(ind))]
    theta_pass.append(theta[ind[o]])
    
    

# plotting the possible theta that dron can access  
  for i,j in zip(theta_pass,r_pass): 
    plt.polar(i, j, 'r+')
  k=k+1
  plt.figure() 
 
```
